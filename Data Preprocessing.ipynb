## Getting the associated Quadrant Number

```python
rankings = pd.read_csv('ACCEDEranking.csv')
rankings
```

```python
#this stores all clipID of all dataset according to quadrant
quadrant_ID_rankings = {
    1 : [],
    2 : [],
    3 : [],
    4 : []
}
```

```python
def determine_quadrant(row):
    valence = row['valenceRank']
    arousal = row['arousalRank']
    midpoint = 4900
    
    if valence >= midpoint:
        if arousal >= midpoint:
            return 1
        else:
            return 4
    else:
        if arousal >= midpoint:
            return 2
        else:
            return 3

rankings['quadrant'] = rankings.apply(determine_quadrant, axis=1)
rankings
```

```python
def get_quadrant_number(clipID, quadrant_IDs):
    number = 0
    for key, value in quadrant_IDs.items():
        if clipID in value:
            number = key
            break
    
    return number
```

```python
quadrant_sizes = rankings['quadrant'].value_counts()
quadrant_sizes
```

```python
# Iterate through the DataFrame and populate the dictionary
for index, row in rankings.iterrows():
    quadrant = row['quadrant']
    clipID = row['clipID']
    quadrant_ID_rankings[quadrant].append(clipID)
```

```python
# Apply the function to create the new column
df['quadrant'] = df['clipID'].apply(lambda x: get_quadrant_number(x, quadrant_ID_rankings))
print_df(df)
```

```python
quadrant_sizes = df['quadrant'].value_counts()
quadrant_sizes
```

## Data Extrapolation

```python
# Verify the shape of the data
print(f"Quadrant 1 data rows: {quadrant_1_df.shape[0]}")
print(f"Quadrant 3 data rows: {quadrant_3_df.shape[0]}")
print(f"Quadrant 4 data rows: {quadrant_4_df.shape[0]}")
print(f"Quadrant 2 data rows: {quadrant_2_df.shape[0]}")
```

```python
# Initial sizes of the quadrants
initial_sizes = {
    'q1': quadrant_1_df.shape[0],
    'q2': quadrant_2_df.shape[0],
    'q3': quadrant_3_df.shape[0],
    'q4': quadrant_4_df.shape[0]
}
```

```python
# Find the key with the highest value
key_with_highest_value = max(initial_sizes, key=initial_sizes.get)
print(key_with_highest_value)

# Find the highest value
target_size = initial_sizes[key_with_highest_value]
print(target_size)
```

```python
# Remove the key with the highest value from the dictionary
del initial_sizes[key_with_highest_value]
```

```python
initial_sizes
```

```python
# Calculate the percentage increase needed to match the target size
percentage_increase = {key: ((target_size - value) / value) * 100 for key, value in initial_sizes.items()}

# Print the percentage increase for each quadrant
for quadrant, increase in percentage_increase.items():
    print(f"Percentage increase for {quadrant}: {increase:.2f}%")
```

```python
import numpy as np
import pandas as pd
from scipy.interpolate import interp1d

def expand_data(data, target_size, columns):
    # Generate x values (assuming data is evenly spaced)
    x = np.arange(len(data))
    
    # Interpolate data
    f = interp1d(x, data, axis=0, kind='linear', fill_value='extrapolate')
    
    # Generate new x values for the expanded dataset
    new_x = np.linspace(0, len(data) - 1, target_size)
    
    # Evaluate the interpolation function at the new x values
    expanded_data = f(new_x)
    
    # Convert to DataFrame and assign column headers
    expanded_df = pd.DataFrame(expanded_data, columns=columns)
    
    return expanded_df
```

```python
perform_on_quadrants = {
    'q1': expand_data(quadrant_1_df.values, target_size, quadrant_1_df.columns),
    'q2': expand_data(quadrant_2_df.values, target_size, quadrant_2_df.columns),
    'q3': expand_data(quadrant_3_df.values, target_size, quadrant_3_df.columns),
    'q4': expand_data(quadrant_4_df.values, target_size, quadrant_4_df.columns)
}
```

```python
df_ex = None
quadrant_1_ex = None
quadrant_2_ex = None
quadrant_3_ex = None
quadrant_4_ex = None


quadrants_extrapolated = {
    'whole': df_ex,
    'q1': quadrant_1_ex,
    'q2': quadrant_2_ex,
    'q3': quadrant_3_ex,
    'q4': quadrant_4_ex
}
```

```python
quadrants_orig = {
    'whole': df,
    'q1': quadrant_1_df,
    'q2': quadrant_2_df,
    'q3': quadrant_3_df,
    'q4': quadrant_4_df
}
```

```python
for key in initial_sizes:
    quadrants_extrapolated[key] = perform_on_quadrants[key]
```

```python
quadrants_extrapolated[key_with_highest_value]  = quadrants_orig[key_with_highest_value]
```

```python
quadrants_extrapolated['whole'] = pd.concat(quadrants_extrapolated.values(), ignore_index=True)
print_df(quadrants_extrapolated['whole'])
```

```python
for key, value in quadrants_extrapolated.items():
    print(f"Shape of DataFrame '{key}': {value.shape}")
```
